```python
# Rename feature columns

num_features = features_df.shape[1]

column_names = ["txId"] + [f"feature_{i}" for i in range(1, num_features)]

features_df.columns = column_names

features_df.head()
```

```python
# Merge features with class labels

df = features_df.merge(classes_df, on="txId")

print("Merged dataset shape:", df.shape)

df.head()
```

```python
df["class"].value_counts()
```

```python
# Remove unknown labels

df = df[df["class"] != "unknown"]

df["class"].value_counts()
```

```python
# Convert labels to binary

df["class"] = df["class"].astype(int)

df["label"] = df["class"].apply(lambda x: 1 if x == 1 else 0)

df["label"].value_counts()
```

```python
df = df.drop(columns=["class"])
```

```python
clean_path = DATA_PROCESSED / "elliptic_clean.csv"

df.to_csv(clean_path, index=False)

print("Clean dataset saved to:", clean_path)
```
